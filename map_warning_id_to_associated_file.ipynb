```python
import pandas as pd

indexed_warnings=pd.read_csv("Wsign-compare_warning_index.csv")

indexed_warnings
```

```python
indexed_warnings["warning_text"]
```

```python
#asserts that all warning texts start with "INFO: From Compiling "
[ text for text in indexed_warnings["warning_text"] if text[:21]!="INFO: From Compiling " ]
```

```python
# assertion that all warning texts contain a reference to a ".cc" file
# loose but likely to be  circumstantially valid
[ text for text in indexed_warnings["warning_text"] if ".cc" not in text ]
```

```python
def read_warning_file_path(warning_text):
    i=21
    warning_file_path=""
    while (warning_file_path[-3:] != ".cc"):
        warning_file_path+=warning_text[i]
        i+=1
    return warning_file_path

id_file_mapping = { i:read_warning_file_path(indexed_warnings["warning_text"][i]) for i in range(len(indexed_warnings)) }

id_file_mapping
```

```python
import json

with open("indexed_warning_files.json","w") as out_file:
    out_file.write(json.dumps(id_file_mapping))
```
